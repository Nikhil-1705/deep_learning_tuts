### HTTP File Compression 
1) Types of compression 
a) Loss Less Compression - Where the compression decompression cycle dosen't alter the data that is recovered. It is same byte to byte 
b) Lossy Compression - After the compression decompression cycle is over, the data never remains the same. Eg - Videos on the web, jpeg 

#### End to End Compression 
Compression of the body of a message, that is done by the server. Will last unchanged till it reaches the clients. Intermediate nodes do not compress it.

### HTTP Cacheing 

Q. What is it? Why is it? 
A. Stores a response against a request and reuses it for subsiquent requests. WHY? Faster responses and reduced server loads 

Types priavte and shared cache 
Private - Stored on the browser (is specific to the client)
Shared - Stored between client and server and stores responses reusable across multiple users — these include proxy caches and managed caches like CDNs and reverse proxies.

responses are still valid(fresh) or stale (expired)

### Autherization in HTTP

The "Basic" authentication scheme used in the diagram above sends the credentials encoded but not encrypted. This would be completely insecure unless the exchange was over a secure connection 

Basic - which transmits credentials as user ID/password pairs, encoded using base64. As the user ID and password are passed over the network as clear text (it is base64 encoded, but base64 is a reversible encoding), the basic authentication scheme is not secure. HTTPS/TLS should be used with basic authentication to prevent credential interception.
