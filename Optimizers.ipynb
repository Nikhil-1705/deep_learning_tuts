**Convex vs Non-Convex Optimization Landscapes**
- **Convex**: Smooth, only one minimum, easy for optimizers to find the best solution.
- **Non-Convex**: Many valleys/hills, many local minima and maxima. Optimizers may get “stuck”, so advanced optimizers/heuristics are needed.

_Note: Most deep learning loss surfaces are highly non-convex!_

3 major problems that non convex function raises 
1. High Curvature 
2. Consistent gradients (sometimes say in the graph of a non convex function there is almost a straight line, so this will affect a normal GD, making it perform slowly)
3. Noisy gradient (There is a chance that there are a lot of local minima's)

### Faster Optimizers

#### ⚡ Momentum Optimization

---

#### ❓ Why is Momentum Needed?

**Gradient Descent (GD)** takes **small, consistent steps** in the direction of the steepest slope.

While this works, it has some limitations:
- **Slow convergence**, especially in long, narrow valleys
- Can **oscillate** in directions where the slope changes rapidly
- Doesn’t consider **past gradients**

👉 In short: **GD only looks at the current step, not the past path**

---

#### 💡 Intuition Behind Momentum

Think of a **ball rolling down a hill**:

- At the start → it moves slowly  
- As it rolls → it **builds speed (momentum)**, say the past 3 gradients point in the same direction, so it will move faster in the 4th direction
- Eventually → it moves faster and more smoothly toward the bottom  

👉 Momentum in optimization works the same way:
- It **accumulates past gradients**
- Helps the model move **faster in the right direction**
- Reduces unnecessary oscillations

---

#### ⚙️ How Gradient Descent Works

Standard GD update:
- Update weights using only the current gradient

\[
\theta = \theta - \eta \nabla_\theta J(\theta)
\]

- **θ** → model parameters  
- **η (eta)** → learning rate  
- **∇J(θ)** → gradient (slope)

👉 Problem: No memory of previous steps

---

#### 🚀 How Momentum Works

Momentum introduces a new concept: **velocity (or momentum vector)**

Instead of using just the current gradient:
- It **remembers past gradients**
- Builds a **running average of gradients**

Steps:
1. Compute gradient
2. Update momentum (velocity)
3. Update weights using this momentum

👉 Key idea:
- Gradient controls **acceleration**
- Momentum controls **speed**

---

#### 🧪 Key Hyperparameter

- **β (beta)** → momentum term
  - Range: **0 to 1**
  - Typical value: **0.9**

Meaning:
- **β = 0** → no momentum (same as GD)
- **β close to 1** → strong momentum (less friction), so imagine a ball rolling w/o frincton, so it won't stop only 

---

#### 📈 Why Momentum Helps

- **Faster convergence**
- **Reduces oscillations**
- Helps move smoothly through **ravines**
- More stable than vanilla GD
